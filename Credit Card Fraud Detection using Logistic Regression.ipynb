#Importing the Libraries/ Dependancies

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

##Loading the dataset to a pandas data frame


```python
credit_card_data = pd.read_csv("/content/creditcard.csv")
```

##Handling missing values

```python
credit_card_data = credit_card_data.dropna()
```

#Check the distribution of Legit transactions and fraudulent transactions

```python
credit_card_data['Class'].value_counts()
```

This Dataset is highly unbalanced

0 --> Normal Transactions

1 --> Fraudulent Transactions

##Separating the Data for analysis

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

##Statistical Measures for the Data

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

##Compare the values for both transactions

```python
credit_card_data.groupby("Class").mean()
```

#Undersampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent transactions --> 213

```python
legit_sample = legit.sample(n=213)
```

##Concatenating two Dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset["Class"].value_counts()
```

```python
new_dataset.groupby("Class").mean()
```

# Splitting the Data into Features and Targets

```python
X = new_dataset.drop("Class", axis=1)
y = new_dataset["Class"]
```

```python
print(X)
```

```python
print(y)
```

#Split the Data into training data and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 2 )
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#Model Training

##Logistic Regression

```python
model = LogisticRegression()
```

##Training the logistic Redgression model with the training Data

```python
model.fit(X_train, y_train)
```

#Model Evaluation

##Accuracy Score

```python
#Accuracy on training Data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print("Accuracy on Training data: ",training_data_accuracy )
```

```python
#Accuracy on Test Data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print("Accuracy on test Data: ", test_data_accuracy )
```
